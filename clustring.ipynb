1. What is unsupervised learning in machine learning?

Unsupervised learning is a machine learning technique where the model learns patterns from unlabeled data without
predefined outputs.

2. How does K-Means clustering work?

K-Means divides data into K clusters by assigning points to the nearest centroid and repeatedly updating centroids
until convergence.

3. Explain dendrogram in hierarchical clustering.

A dendrogram is a tree-like diagram that shows how clusters are merged or divided in hierarchical clustering.

4. Difference between K-Means and Hierarchical Clustering.

K-Means requires the number of clusters beforehand, while Hierarchical Clustering creates a hierarchy of clusters
without requiring K initially.

5. Advantages of DBSCAN over K-Means.

DBSCAN can detect arbitrary-shaped clusters and identify noise points, unlike K-Means.

6. When is Silhouette Score used?

It is used to evaluate clustering quality by measuring how similar a point is to its own cluster compared to other
clusters.

7. Limitations of Hierarchical Clustering.

It is computationally expensive and difficult to scale for large datasets.

8. Why is feature scaling important in clustering?
Scaling ensures all features contribute equally because clustering algorithms rely on distance calculations.

9. How does DBSCAN identify noise points?
Points not belonging to dense regions are labeled as noise or outliers.

10. Define inertia in K-Means.

Inertia is the sum of squared distances between data points and their assigned cluster centroids.

11. What is the elbow method?


It is a technique to find the optimal K value by plotting inertia against the number of clusters.

12. Describe density in DBSCAN.

Density refers to the concentration of points within a neighborhood region.

13. Can hierarchical clustering work on categorical data?

Yes, by using appropriate distance measures such as Hamming distance.

14. What does a negative Silhouette Score indicate?

It indicates that samples may have been assigned to the wrong cluster.

15. Explain linkage criteria.

Linkage criteria determine how distances between clusters are calculated during hierarchical clustering.

16. Why does K-Means perform poorly on varying densities?

Because it assumes clusters are spherical and evenly distributed.

17. Core parameters in DBSCAN.

The main parameters are eps (radius) and min_samples (minimum points required to form a dense region).

18. How does K-Means++ improve initialization?

It selects initial centroids more carefully to improve convergence and reduce poor clustering.

19. What is agglomerative clustering?

It is a bottom-up hierarchical clustering method where each point starts as its own cluster.

20. Why is Silhouette Score better than inertia?

Silhouette Score measures both cohesion and separation, while inertia only measures compactness

Code 2: Generate overlapping clusters and apply KMeans with decision boundaries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
X, y = make_blobs(n_samples=300, centers=3,
                  cluster_std=2.5, random_state=42)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
h = 0.02
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.show()
